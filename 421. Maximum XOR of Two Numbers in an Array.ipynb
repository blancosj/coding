<a href='https://leetcode.com/problems/maximum-xor-of-two-numbers-in-an-array/'>421. Maximum XOR of Two Numbers in an Array</a>

Given a non-empty array of numbers, a0, a1, a2, … , an-1, where 0 ≤ ai < 231.

Find the maximum result of ai XOR aj, where 0 ≤ i, j < n.

Could you do this in O(n) runtime?

__Example:__
```
Input: [3, 10, 5, 25, 2, 8]

Output: 28

Explanation: The maximum result is 5 ^ 25 = 28.
```

````java
public class Solution {
    public int findMaximumXOR(int[] nums) {
        int max = nums[0];
        for (int num : nums) max = Math.max(num, max);
        int curMax = max, l = 0;
        while (curMax > 0) {
            curMax >>= 1;
            l++;
        }
        int maxXor = 0, curXor = 0;
        Set<Integer> prefixes = new HashSet<>();
        for (int i = 1; i <= l; ++i) {
            prefixes.clear();
            maxXor <<= 1;
            curXor = maxXor | 1;
            for (int num : nums) {
                int prefix = num >> l - i;
                if (prefixes.contains(curXor ^ prefix)) {
                    maxXor = curXor;
                    break;
                }
                prefixes.add(num >> l - i);
            }
        }
        return maxXor;
    }
}
````

````java
public class Solution {
    public int findMaximumXOR(int[] nums) {
        int max = nums[0];
        for (int num : nums) max = Math.max(num, max);
        int l = Integer.toBinaryString(max).length();
        int n = nums.length;
        String[] strNums = new String[n];
        int bitmask = 1 << l;
        // padding left
        for (int i = 0; i < n; ++i) {
            strNums[i] = Integer.toBinaryString(bitmask | nums[i]).substring(1);    
        }
        TrieNode trie = new TrieNode();
        int maxXor = 0;
        for (String num : strNums) {
            TrieNode node = trie, xorNode = trie;
            int curXor = 0;
            for (char bit : num.toCharArray()) {
                // insert new node
                if (node.children.containsKey(bit)) {
                    node = node.children.get(bit);
                } else {
                    TrieNode newNode = new TrieNode();
                    node.children.put(bit, newNode);
                    node = newNode;
                }
                char toggledBit = bit == '1' ? '0' : '1';
                if (xorNode.children.containsKey(toggledBit)) {
                    curXor = (curXor << 1) | 1;
                    xorNode = xorNode.children.get(toggledBit);
                } else {
                    curXor = curXor << 1;
                    xorNode = xorNode.children.get(bit);
                }
            }
            maxXor = Math.max(curXor, maxXor);
        }
        return maxXor;
    }
    private class TrieNode {
        public Map<Character, TrieNode> children = new HashMap<>();
    }
}
````
