```java
public double maxProfit(double[] prices) {
    int len = prices.length;
    double minPrice = Integer.MAX_VALUE;
    double maxProfit = 0;
    for (int i = 0; i < len; ++i) {
        if (minPrice > prices[i]) {
            minPrice = prices[i];
        } else if (prices[i] - minPrice > maxProfit) {
            maxProfit = prices[i] - minPrice;
        }
    }
    return maxProfit;
}
```
